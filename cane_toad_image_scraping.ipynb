```
from bs4 import BeautifulSoup

import requests
import urllib.request
import shutil
```

```
url = "https://www.feralscan.org.au/toadscan/pagecontent.aspx?page=toad_toadsandnativefrogs"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

images = soup.find_all("div", class_='fsc-image-row card')

images
```

```
cane_toad_image_info = []
native_frog_image_info = []

for image in images:
    image_tag = image.findChildren("img")
    if "Cane" in image_tag[0]["src"]:
      cane_toad_image_info.append((image_tag[0]["src"]))
    elif "Native" in image_tag[0]["src"]:
      native_frog_image_info.append((image_tag[0]["src"]))

display(cane_toad_image_info)
display(native_frog_image_info)
```

```
cane_toad_image_urls = []
native_frog_image_urls = []

for i in range(len(cane_toad_image_info)):
  path = cane_toad_image_info[i]
  full_path = 'https://www.feralscan.org.au/' + path
  cane_toad_image_urls.append(full_path)

for i in range(len(native_frog_image_info)):
  path = native_frog_image_info[i]
  full_path = 'https://www.feralscan.org.au/' + path
  native_frog_image_urls.append(full_path)

display(cane_toad_image_urls)
display(native_frog_image_urls)
```

```
import os
import pathlib

cane_toad_data_dir = os.getcwd() + '/drive/My Drive/Colab Notebooks/Datasets/cane_toad_classification/cane_toads/'
native_frog_data_dir = os.getcwd() + '/drive/My Drive/Colab Notebooks/Datasets/cane_toad_classification/native_frogs/'

def download_image(directory, image, index):
    response = requests.get(image, stream=True)
    
    file = open(directory + "{}.jpg".format(index), 'wb')
    print("Downloading " + image)
    
    response.raw.decode_content = True
    shutil.copyfileobj(response.raw, file)
    del response
```

```
for i in range(0, len(cane_toad_image_urls)):
    download_image(cane_toad_data_dir, cane_toad_image_urls[i], i)

for i in range(0, len(native_frog_image_urls)):
    download_image(native_frog_data_dir, native_frog_image_urls[i], i)
```
